The question that i want to answer is Which diagnosis groups show the highest readmission density?
and i used these variables: diagnosis_group, readmitted_30d, age

and here is the Axes' details:

X: diagnosis_group

Y: readmitted_30d

Hue: age group

Chart: Bar plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


data=pd.read_csv("synthetic_healthcare_1700_clean.csv")
df = pd.DataFrame(data)
```

```python


bins = [0, 30, 50, 70, 101] 
labels = ['<30', '30-50', '50-70', '70+']

df['age group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# 3. Plotting
plt.figure(figsize=(10, 6))
sns.barplot(
    data=df, 
    x='Diagnosis Group', 
    y='Readmitted (30 Days)', 
    hue='age group'
)

plt.title('Readmission Rate by Diagnosis Group and Age Group')
plt.xlabel('Diagnosis Group')
plt.ylabel('Readmission Rate (30d)')
plt.legend(title='Age Group')

plt.tight_layout()
plt.show()
```
