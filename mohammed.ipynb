The question that i want to answer is Which diagnosis groups show the highest readmission density?
and i used these variables: diagnosis_group, readmitted_30d, age

and here is the Axes' details:

X: diagnosis_group

Y: readmitted_30d

Hue: age group

Chart: Bar plot

```python
import pandas as pd # importing the pandas
import seaborn as sns # ipmorting seaborn lib
import matplotlib.pyplot as plt # importing matblotlib


data=pd.read_csv("synthetic_healthcare_1700_clean.csv") # reading CSV file
df = pd.DataFrame(data) # change CSV to DataFrame
```

```python
bins = [0, 30, 50, 70, 101] # to determin the bins for the graph
labels = ['<30', '30-50', '50-70', '70+'] # to determine the labale that are used in the graph

df['age group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False) # this line to 

# 3. Plotting the graph
plt.figure(figsize=(10, 6))
sns.barplot(
    data=df, 
    x='Diagnosis Group', # x-axis
    y='Readmitted (30 Days)', # y-axis
    hue='age group' # this to determine the color based on Age-group
)

# adding some details; like adding the title for the 
plt.title('Readmission Rate by Diagnosis Group and Age Group')
plt.xlabel('Diagnosis Group')  # Xlabel title
plt.ylabel('Readmission Rate (30d)') # Ylable title
plt.legend(title='Age Group') 

plt.tight_layout() 
plt.show() # to show the figure
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('synthetic_healthcare_1700_clean.csv', sep=',')

stackedBar = df.groupby(["Region", "Gender"]).size().unstack()
# Group by Region and Gender, count the occurrence and then unstack for plotting


# Stacked bar chart

chart = stackedBar.plot(
    kind="bar",
    stacked=True,
    figsize=(12,8) # Figure size of chart
)

# Adding labels
for c in chart.containers:
    chart.bar_label(c, label_type="center")

sns.set_style('white') # Background color to white

plt.title("Patients by Gender and Region") # Set the Title of the chart
plt.xlabel("Region") # Set label of the X-Axis
plt.ylabel("Number of Patients") # Set label of the Y-Axis
plt.show() # Build and show the plot
```
